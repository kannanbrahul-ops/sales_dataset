```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r'C:\Users\ACER\Downloads\sales_dataset.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Region']=df['Region'].fillna('Sales')
df
```

```python
df['Product']=df['Product'].fillna(df['Product'].mode()[0])
df
```

```python
df['Product']=df['Product'].fillna('UnitPrice')
df
```

```python
df['Quantity']=df['Quantity'].replace('five','5')
df
```

```python
df['Quantity']=pd.to_numeric(df['Quantity'])
df
```

```python
df.dtypes
```

```python
df['Quantity']=pd.to_numeric(df['Quantity'].fillna(df['Quantity'].mode()[0]))
df
```

```python
df['UnitPrice']=df['UnitPrice'].replace('unknown','4')
df
```

```python
df['UnitPrice']=pd.to_numeric(df['UnitPrice'])
df
```

```python
df['OrderDate']=pd.to_datetime(df['OrderDate'])
df
```

```python
df['UnitPrice']=df['UnitPrice'].replace(4,df['UnitPrice'].mean())
df
```

```python
df.isnull().sum()
```

```python
df['UnitPrice']=df['UnitPrice'].fillna(df['UnitPrice'].mean())
df
```

```python
df.isnull().sum()
```

```python
data=df.groupby('Region')['TotalAmount'].sum().reset_index()
data
```

```python
sns.barplot(x='Region',y='TotalAmount',data=data,hue='Region')
plt.show()
```

```python
data1=df.groupby('Product')['TotalAmount'].sum().reset_index()
data1
```

```python
sns.barplot(x='Product',y='TotalAmount',data=data1,hue='Product')
plt.show()
```

```python
data2=df.groupby('SalesPerson')['Quantity'].sum().reset_index()
data2
```

```python
sns.barplot(x='SalesPerson',y='Quantity',data=data2,hue='SalesPerson')
plt.show()
```
